```python
sns.barplot(x =customerchurn["Contract"].value_counts().keys(),y=customerchurn["Contract"].value_counts().values, data = customerchurn,color="orange")
plt.xlabel("Contract Type of customer")
plt.ylabel("Count")
plt.title("Distribution of Contract")
plt.show()
```

a. Set the x-axis label to be â€˜Tenure of the customerâ€™<br>
