1. Sam has to build a bar-plot for the ‘Contract’ column<br>
a. Set the x-axis label to be ‘Contract Type of customer’<br>
b. Set the y-axis label to be ‘Count’<br>
c. Set the title of the plot to be ‘Distribution of Contract’<br>
d. Assign ‘orange’ color to all the bars

```python
customerchurn["Contract"].value_counts()
```

```python
customerchurn["Contract"].value_counts().values
```

```python
sns.barplot(x =customerchurn["Contract"].value_counts().keys(),y=customerchurn["Contract"].value_counts().values, data = customerchurn,color="orange")
plt.xlabel("Contract Type of customer")
plt.ylabel("Count")
plt.title("Distribution of Contract")
plt.show()
```

3. Sam has to build a scatter-plot between ‘TotalCharges’ & ‘tenure’. ‘TotalCharges’ should be on
the y-axis and ‘tenure’ should be on the x-axis<br>
a. Set the x-axis label to be ‘Tenure of the customer’<br>
b. Set the y-axis label to be ‘Total charges Incurred’<br>
c. Set the title of the plot to be ‘Total Charges vs Tenure’<br>
d. Assign ‘indigo’ color to the points

```python
sns.scatterplot(x="tenure",y= "TotalCharges",size=10,data=customerchurn,palette="indigo")
plt.xlabel("Tenure of the customer")
plt.ylabel("Total charges Incurred")
plt.title("Total Charges vs Tenure")
plt.show()
```

4. Sam has to build a box-plot between ‘MonthlyCharges’ & ‘PaymentMethod’. ‘MonthlyCharges’
should be on the y-axis and ‘PaymentMethod’ should be on the x-axis.<br>
a. Set the x-axis label to be ‘Payment Method of customer’<br>
b. Set the y-axis label to be ‘Monthly Charges Incurred’<br>
c. Set the title of plot to be ‘Monthly Charges vs Payment Method’<br>
d. Assign ‘olive’ color to the box-plots

```python
sns.boxplot(x = "PaymentMethod",y= "MonthlyCharges",data=customerchurn,color="olive")
plt.xlabel("Payment Method of customer")
plt.ylabel("Monthly Charges Incurred")
plt.title("Monthly Charges vs Payment Method")
plt.show()
```
